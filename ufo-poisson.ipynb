```r
x <- read.csv("regdata.csv", as.is=TRUE)
x
```

```r
library(outreg)
```

```r
m1 <- glm(freq ~ state_code-1, data=x, offset=log(x$pop), family=poisson)
m2 <- glm(freq_30min ~ state_code-1, data=x, offset=log(x$pop), family=poisson)
m3 <- glm(freq_circle ~ state_code-1, data=x, offset=log(x$pop), family=poisson)
```

```r
outreg(list("all"=m1, "30min"=m2, "circle"=m3))
```

```r
parse_glm <- function(m) {
  s <- summary(m)$coefficients
  states <- gsub("state_code", "", row.names(s))
  coefs <- s[,1]
  ses <- s[,2]
  o <- data.frame(state=states, coef=coefs, se=ses)
  
  # coefs are log(mean)
  # to covert it to log10, we multiply log10(e)
  o$coef <- o$coef * log10(exp(1))
  row.names(o) <- NULL
  o
}
o1 <- parse_glm(m1)
o1$model <- "all"

o2 <- parse_glm(m2)
o2$model <- "30min"

o3 <- parse_glm(m3)
o3$model <- "circle"

o <- rbind(o1, o2, o3)
o
```
